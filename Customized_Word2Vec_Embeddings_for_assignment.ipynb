```
def Softmax(np_list_inp):
  ''' 
    This function takes a numpy list as input that will contains Z_js. This data structure's name is np_list_inp.
    The function will return the softmax output of each entry as a form of numpy list. This data structure's name is np_list_out.

    Using instructions of the function :
      Consider the function has 10 inputs or values and we need Softjax of 4th element i.e. Z_4,
      so 4 element in np_list_out would be sigma(Z_4) i.e. softmax of Z_4
  '''
  np_list_inp = np.array(np_list_inp)  # In case the input vector is not an numpy array but a simple list
  exp_list = np.exp(np_list_inp) 
  #exp_list = np.exp(np_list_inp - np.max(np_list_inp))   # This version is complex but numerically more stable.
  sum_of_exp = sum(exp_list)
  
  return exp_list / sum_of_exp
```

```
# The adjusted_sampling function will be used to get the probabilities for picking up a random negative sample.

def adjusted_sampling(word_count):
  word_choice = {}
  sum_p = 0.00

  for i in word_count:
    word_choice[i] = word_count[i] ** 0.75
    sum_p += word_choice[i]

  for i in word_choice:
    word_choice[i] /= sum_p

  return word_choice
```

```
def negative_samples():
  """
    This function returns num_neg samples from the corpus for Negative Sampling
  """
  rand = np.random.choice(list(word_choice.keys()), size=num_neg, p=list(word_choice.values()))
  samples = []

  for i in rand:
    samples.append(np.array(word2onehot(i)))

  return np.array(samples)
```

```
def skg(inp, labels, w1, w2, loss, learning_rate):
  """
      inp -> Onehot vector of the input or the target word in nunmpy array format.
      labels -> list of all the onehot vectors of the context words.
  """
  h = np.dot(w1.T, inp)
  u = np.dot(w2.T, h)
  y_pred = Softmax(u)
 
  error = np.array([-label + y_pred.T for label in labels])
  dw2 = np.outer(h, np.sum(error, axis=0))
  dw1 = np.outer(inp, np.dot(w2, np.sum(error, axis=0)))
 
  new_W1, new_W2 = SGD(w1,w2,dw1,dw2, learning_rate)
 
  loss += -np.sum([u[label == 1] for label in labels]) + len(labels) * np.log(np.sum(np.exp(u)))
 
  return new_W1, new_W2, loss
```

```
def cbow(inp, label, w1, w2, loss, learning_rate):
  """
    inp -> list of onehot vectors of context words
    label -> onehot vector of centre word.
  """
  
  x = np.mean(inp, axis=0)
  h = np.dot(w1.T, x)
  u = np.dot(w2.T, h)
  y_pred = Softmax(u)
 
  error = -label + y_pred
  dw2 = np.outer(h, error)
  dw1 = np.outer(x, np.dot(w2, error))
 
  new_W1, new_W2 = SGD(w1,w2,dw1,dw2, learning_rate)
 
  loss += -float(u[label == 1]) + np.log(np.sum(np.exp(u)))
 
  return new_W1, new_W2, loss
```

```
def sigmoid(x):
  return 1 / (1 + np.exp(-x))

def nsskg(inp, label, w1, w2, rate):
  h = np.dot(w1.T, inp)
  u = np.dot(w2.T, h)

  neg_words = negative_samples()

  u = sigmoid(u)
  mask = inp

  for i in neg_words:
    mask += i

  u = np.dot(u, mask)
  u = u - label

  dw1 = np.dot(w1.T, u)
  dw1 = np.outer(inp, dw1)
  dw2 = np.outer(h, u)

  return SGD(w1, w2, dw1, dw2, rate)
```

```
def SGD(w1,w2,dw1,dw2, learning_rate):
  """
    Each epoch or iteration accounts for updation due single training cycle.
  """
  new_w1 = w1 - (learning_rate * dw1)
  new_w2 = w2 - (learning_rate * dw2)

  return new_w1, new_w2
```

```
vector_size = [200,300, 350, 400, 500]
learning_rate = [0.001, 0.01, 0.1, 0.5]
 
for size in vector_size: 
  print("Starting the training with embedding size = ", size, "\n\n")
  np.random.seed(100)
  W1 = np.random.rand(v_count, size)
  W2 = np.random.rand(size, v_count)
  loss = 0.00
 
# For SKG
  if (choice == "SKG"):
    for rate in learning_rate: 
      itr = 0
      print("Starting the training with learning rate = ", rate, "\n\n")
      for itr in range(epochs):
        
        for sentence_skg in corpus:
          sent_skg_len = len(sentence_skg)
          for i_skg, word_skg in enumerate(sentence_skg):
            w_skg_target = word2onehot(sentence_skg[i_skg])
            inp = np.array(w_skg_target)
 
            w_skg_context = []
            for j in range(i_skg - window, i_skg+ window+1):
              if j!=i_skg and j<= sent_skg_len - 1 and j>=0:
                w_skg_context.append(word2onehot(sentence_skg[j]))
            
            labels = np.array(w_skg_context)
            W1, W2, loss = skg(inp, labels, W1, W2, loss, rate)
 
            del w_skg_context
            del w_skg_target
            del labels
            del inp
        
        print(" Result after epoch number ",itr, ": loss = ", loss, " for rate, size = ", rate, "  ",size)
        print("\n\n\n --------------------------------------------------------------------------------------------------------------------------------------")
    
      path = "/Jash/My Drive/NLP/" + "Window_" + str(window) + "/" + str(choice) + "/" + str(rate) + "_" + str(size) + "weight_numpy" 
      np.save(path, W1)
 
# For CBOW
  elif (choice == "CBOW"):
    for rate in learning_rate: 
      print("Starting the training with learning rate = ", rate, "\n\n")
      for itr in range(epochs):
        for sentence_cbow in corpus:
          sent_cbow_len = len(sentence_cbow)
 
          for i_cbow, word_cbow in enumerate(sentence_cbow):
            w_cbow_target = word2onehot(sentence_cbow[i_cbow])
            label = np.array(w_cbow_target)
 
            w_cbow_context = []
            for j in range(i_cbow - window, i_cbow + window + 1):
              if j!=i_cbow and j<=sent_cbow_len-1 and j>=0:
                w_cbow_context.append(word2onehot(sentence_cbow[j]))
            
            inp = np.array(w_cbow_context)
            W1, W2, loss = cbow(inp, label, W1, W2, loss, rate) 
 
            del w_cbow_context
            del w_cbow_target
            del label
            del inp
 
        print(" Result after epoch number ",itr, ": loss = ", loss, " for rate, size = ", rate, "  ",size)
        print("\n\n\n --------------------------------------------------------------------------------------------------------------------------------------")
    

      path = "/Jash/My Drive/NLP/" + "Window_" + str(window) + "/" + str(choice) + "/" + str(rate) + "_" + str(size) + "weight_numpy" 
      np.save(path, W1)

# For Negative Samapling SKG
  elif (choice == "NSSKG"):
    print('NSSKG')
    for rate in learning_rate: 
      print("Starting the training with learning rate = ", rate, "\n\n")
      for itr in range(epochs):
        for sentence_ns in corpus:
          sent_ns_len = len(sentence_ns)

          for i_ns, word_ns in enumerate(sentence_ns):
            inp_a = word2onehot(sentence_ns[i_ns])
            inp = np.array(inp_a)

            for i in range(i_ns-window, i_ns + window + 1):
              if i != i_ns and i < sent_ns_len and i >= 0:

                label_a = word2onehot(sentence_ns[i])
                label = np.array(label_a)
                W1, W2 = nsskg(inp, label, W1, W2, rate)

            del label
            del inp
            del label_a
            del inp_a

        print(" Result after epoch number ",itr, ": loss = ", loss, " for rate, size = ", rate, "  ",size)
        print("\n\n\n --------------------------------------------------------------------------------------------------------------------------------------")

      path = "/Jash/My Drive/NLP/" + choice + str(rate) + "_" + str(size) + "weight_numpy" 
      np.save(path, W1)
 
  else:
    print(" You entered wrong information !!!! ")
```
